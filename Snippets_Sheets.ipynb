# Importing data from Google Sheets

```
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import numpy as np
import time
model = YOLO("yolov8n.pt")
cap = cv2.VideoCapture('person3.mp4')  # مسیر درست فایل ویدیو

if not cap.isOpened():
    print("ویدیو باز نشد!")
else:
    print("ویدیو بارگذاری شد")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # اجرای مدل YOLO روی فریم
    results = model(frame)

    annotated_frame = results[0].plot()

    # تبدیل به RGB برای نمایش در Colab
    rgb_frame = cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB)


    plt.imshow(rgb_frame)
    plt.axis('off')
    plt.show()

    break  # اینو بردار اگه بخوای کل ویدیو رو نمایش بدی

cap.release()
```
